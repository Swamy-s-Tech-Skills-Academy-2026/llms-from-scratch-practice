# Chapter 4: Practice Lab - Performance and Parameters

I want to verify my intuition about how many parameters exist in standard model configurations by writing a script that counts them. This is the hands-on simulation corresponding to my reading notes on FLOPs and parameter counting.

```
import torch
import torch.nn as nn

# Ensures reproducibility per the learning guidelines
torch.manual_seed(42)
print("Ready for performance analysis!")
```

## 1. Defining a Tiny Model to Profile

I will create a very basic mock FeedForward layer as a proxy for the GPT network to see how parameters scale.

```
class MockLayer(nn.Module):
    def __init__(self, emb_dim):
        super().__init__()
        self.fc1 = nn.Linear(emb_dim, 4 * emb_dim)
        self.fc2 = nn.Linear(4 * emb_dim, emb_dim)

    def forward(self, x):
        return self.fc2(self.fc1(x))

model = MockLayer(768)
```

## 2. Counting the Parameters

I can loop through `model.parameters()` and multiply the shapes to get the raw parameter count.

```
total_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print(f"Total Parameters in MockLayer: {total_params:,}")

# Memory calculation (fp32 = 4 bytes per parameter)
mem_bytes = total_params * 4
mem_mb = mem_bytes / (1024 ** 2)
print(f"Memory strictly for weights: {mem_mb:.2f} MB")
```
